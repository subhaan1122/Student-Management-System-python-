```python
students = []
total = 0

def enter():
    global total
    choice = int(input("How many students do you want to enter? "))

    for i in range(choice):
        print(f"\nEnter the data of student {total + 1}")
        name = input("Enter name: ")
        roll_no = input("Enter Roll no: ")
        department = input("Enter department: ")
        course = input("Enter course: ")
        students.append({
            "name": name,
            "roll_no": roll_no,
            "department": department,
            "course": course
        })
        total += 1

def show():
    if total == 0:
        print("No data is entered.")
    else:
        for i, student in enumerate(students):
            print(f"\nData of Student {i + 1}")
            print(f"Name: {student['name']}")
            print(f"Roll no: {student['roll_no']}")
            print(f"Department: {student['department']}")
            print(f"Course: {student['course']}")

def search():
    if total == 0:
        print("No data is entered.")
    else:
        roll_no = input("Enter the roll no of the student: ")
        found = False
        for student in students:
            if student["roll_no"] == roll_no:
                print(f"Name: {student['name']}")
                print(f"Roll no: {student['roll_no']}")
                print(f"Department: {student['department']}")
                print(f"Course: {student['course']}")
                found = True
        if not found:
            print("Student not found.")

def update():
    if total == 0:
        print("No data is entered.")
    else:
        roll_no = input("Enter the roll no of the student you want to update: ")
        for student in students:
            if student["roll_no"] == roll_no:
                print("\nPrevious data:")
                print(f"Name: {student['name']}")
                print(f"Roll no: {student['roll_no']}")
                print(f"Department: {student['department']}")
                print(f"Course: {student['course']}")

                print("\nEnter new data:")
                student["name"] = input("Enter name: ")
                student["roll_no"] = input("Enter Roll no: ")
                student["department"] = input("Enter department: ")
                student["course"] = input("Enter course: ")
                print("Record updated successfully!")
                return
        print("Student not found.")

def delete_record():
    global total
    if total == 0:
        print("No data is entered.")
    else:
        print("Press 1 to delete all records.")
        print("Press 2 to delete a specific record.")
        choice = int(input())

        if choice == 1:
            students.clear()
            total = 0
            print("All records deleted.")
        elif choice == 2:
            roll_no = input("Enter the roll no of the student you want to delete: ")
            for i, student in enumerate(students):
                if student["roll_no"] == roll_no:
                    students.pop(i)
                    total -= 1
                    print("Record deleted successfully!")
                    return
            print("Student not found.")
        else:
            print("Invalid input.")

def main():
    while True:
        print("\n1. Enter data")
        print("2. Show data")
        print("3. Search data")
        print("4. Update data")
        print("5. Delete data")
        print("6. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            enter()
        elif choice == 2:
            show()
        elif choice == 3:
            search()
        elif choice == 4:
            update()
        elif choice == 5:
            delete_record()
        elif choice == 6:
            print("Exiting...")
            break
        else:
            print("Invalid input!")

if __name__ == "__main__":
    main()
```
